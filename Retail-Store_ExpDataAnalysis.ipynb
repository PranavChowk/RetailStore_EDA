 <li>Office Supplies can be accounted for having high discount rates for having corporates as their main consumers</li>
 <li>Technology has the highest sales as well as profit with phones being sold the most and copiers earning the highest profit</li>
    <li>Selling tables need to be focused on as they are the second highest sold with most loss.</li>
    <li>Consumers segment products are sold the highest with high profits as well as high discounts</li>
